```python
import qrcode


```

```python
upi_id = input("Enter your UPI ID = ")
```

```python
#upi://pay?pa=UPI_ID&pn=NAME&am=Amount&cu=CURRENCY&tn=MESSAGE
#Defining the payment URL based on the UPI ID and the payment app
```

```python
googlepay_url = f'upi://pay?pa={upi_id}&pn=Receipient%20Name&mc=1234'
paytm_url = f'upi://pay?pa={upi_id}&pn=Receipient%20Name&mc=1234'
```

```python
#create QR code for each payment app
```

```python
googlepay_qr = qrcode.make(googlepay_url)
paytm_qr = qrcode.make(paytm_url)
```

```python
#saving the QR code to image file
```

```python
googlepay_qr.save('googlepay_qr.png')
```

```python
paytm_qr.save('paytm_qr.png')
```

```python
#Display the QR codes
```

```python
googlepay_qr.show()
```
